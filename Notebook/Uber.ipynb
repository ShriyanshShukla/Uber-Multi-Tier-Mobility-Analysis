## Making Date and Time

```python
df = df.copy()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Time'] = pd.to_datetime(df['Time'], format="%H:%M:%S").dt.time
```

```python
df.info()
```

## Handling Nulls

```python
df.isnull().sum()
```

```python
df = df.drop(['Incomplete Rides','Incomplete Rides Reason'],  axis=1)
```

```python
df.head()
```

## Feature Engineering

#### Creating Time Bucket Column

```python
def bucket(t):
    if t < datetime.time(6, 0):
        return '12 AM - 6 AM'
    elif t < datetime.time(12, 0):
        return '6 AM - 12 PM'
    elif t < datetime.time(18, 0):
        return '12 PM - 6 PM'
    else:
        return '6 PM - 12 AM'
```

```python
df['Time Bucket'] = df['Time'].apply(bucket)
```

```python
df.head()
```

#### Creating Day/Night Column

```python
def day_night(t):
    if t < datetime.time(6,0):
        return 'Night'
    elif t < datetime.time(18,0):
        return 'Day'
    else:
        return 'Night'
```

```python
df['Day/Night'] = df['Time'].apply(day_night)
```

```python
df.head()
```

#### Creating Weekday Column

```python
day = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']

df['Weekday'] = df['Date'].dt.weekday.apply(lambda x: day[x])
```

```python
df.head()
```

#### Creating Estimated Revenue Column

```python
total_revenue = df['Booking Value'].sum()
total_revenue
```

```python
average_value = total_revenue / df['Booking Value'].count()
average_value = round(average_value)
average_value
```

```python
df['Estimated Revenue'] = df['Booking Value'].fillna(average_value)
```

```python
df.head()
```

#### Creating Payment Type column

```python
df['Payment Type'] = df['Payment Method'].replace({
    'UPI':'Digital',
    'Credit Card':'Digital',
    'Debit Card':'Digital',
    'Wallet':'Digital',
    'Cash':'Cash'
})
```

```python
df.head()
```

```python
df.to_csv('Uber_Bookings.csv', index=False)
```
